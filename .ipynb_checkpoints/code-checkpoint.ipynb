```python
import requests
import time
import csv
import sys
from bs4 import BeautifulSoup
import requests

url = "https://www.capterra.com/business-intelligence-software//"
page = requests.get(url)
soup = BeautifulSoup(page.text,"html.parser")
#print(soup.prettify())
```

```python
names = soup.find_all('div', class_ = 'grid')
description = []
url = []
houses = []
for i in names:
    #description.append(i.text)
    title = i.text
    houses.append(title)    
print(houses)
```

```python
names = soup.find_all('h2', class_ = 'listing-name')
full_name = []
full_url = []
for i in names:
    name = i.text.strip()
    url = i.a['href']
    complete = [name, url]
    full_name.append(complete)
print(full_name)
```

```python
names = soup.find_all('div', class_ = 'grid-slim')
full_name = []
full_url = []
for i in names:
    url = i.a['href'] 
    full_name.append('https://www.capterra.com' + url)
print(full_name)
```

```python
names = soup.find_all('div', class_ = 'listing-thumbnail-container')
full_name = []
full_url = []
for i in names:
    name = i.text.strip()
    url = i.a['href']  
    full_name.append(url)
print(full_name)
```

```python
with open ('newtownabbeyrentalproperties.csv','w') as file:
    writer=csv.writer(file)
    writer.writerow(['Address', 'Price'])
    for row in houses:
        writer.writerow(row)
```
